```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop_duplicates(keep = 'first',inplace = True)
```

```python
df.isna().sum().sort_values(ascending = False)
```

```python
  percent=df.isna().sum()/df.shape[0]*100
  result=pd.DataFrame({"columns": percent.keys(), "na%": percent})
  result.reset_index(drop = True,inplace = True)
  result
```

```python
df = df.drop(columns = ['house_rules','license','last review','reviews per month','availability 365'])
```

```python
df['NAME'] = df['NAME'].fillna(value = 'blank')
df['host name'] = df['host name'].fillna(value = 'blank')
df['country'] = df['country'].fillna(value = 'United States')
df['country code'] = df['country code'].fillna(value = 'US')
df['host_identity_verified'] = df['host_identity_verified'].fillna(value = 'unconfirmed')
```

```python
df.isna().sum().sort_values(ascending = False)
```

```python
  percent=df.isna().sum()/df.shape[0]*100
  result=pd.DataFrame({"columns": percent.keys(), "na%": percent})
  result.reset_index(drop = True,inplace = True)
  result
```

```python
#since the missing value are less than 1%, we can just delete it
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df['price']=df['price'].str.replace("$","").str.replace(",","").astype(float)
```

```python
df['service fee']=df['service fee'].str.replace("$","").str.replace(",","").str.replace(" ","").astype(float)
```

```python
df['neighbourhood group']=df['neighbourhood group'].str.replace("brookln","Brooklyn")
```

```python
matrix = df.corr().round(2)
```

```python
plt.figure(figsize=(17, 10))
sns.heatmap(matrix, annot=True,vmax=1, vmin=-1, center=0, cmap = 'vlag').set_title('Correlation Heatmap',fontdict={'fontsize':15},pad=12)
plt.savefig('/Users/meiyeechong/Downloads/heatmap.png')
plt.show()
```
